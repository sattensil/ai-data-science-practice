```
# load preprocessed data
with open("BRS/preprocessed.json", "r") as fp:
  preprocessed_combined = json.load(fp)
```

```
# convert to numeric
tfidf_vectorizer = TfidfVectorizer(stop_words='english', binary=True)
tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_data_with_emails)
```

```
# Create a TruncatedSVD object with 3000 components
svd = TruncatedSVD(n_components = 3000)

# Fit the TruncatedSVD model to the TF-IDF matrix
svd.fit(tfidf_matrix)

# Transform the TF-IDF matrix using TruncatedSVD
transformed_matrix = svd.transform(tfidf_matrix)

svd.explained_variance_ratio_.sum()
```

```
# Create an instance of HDBSCAN
clusterer = cuml.cluster.hdbscan.HDBSCAN(min_cluster_size=2, min_samples = 2)

# Fit the model to your data
clusterer.fit(transformed_matrix)
labels = clusterer.labels_
```
